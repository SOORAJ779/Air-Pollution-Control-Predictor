```python
from sklearn.model_selection import train_test_split
x = dt[['CO\n(mg/m³)', 'Ozone (µg/m³)', 'NO\n(µg/m³)',
       'NO2\n(µg/m³)', 'NOX\n(µg/m³)', 'NH3\n(µg/m³)', 'SO2\n(µg/m³)',
       'PM2.5\n(µg/m³)', 'PM10\n(µg/m³)']]
y = dt['AQI']
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
scaledx = scale.fit_transform(x)
print(scaledx)
```

KNN

```python
from sklearn.model_selection import train_test_split
x = scaledx
y = dt['AQI']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

DECISION TREE

```python
from sklearn.model_selection import train_test_split
x = scaledx
y = dt['AQI']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt1 = DecisionTreeRegressor()
dt1.fit(x_train,y_train)
y_pred = dt1.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

RANDOM FOREST

```python
from sklearn.model_selection import train_test_split
x = scaledx
y = dt['AQI']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor()
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

LINEAR REGRESSION

```python
from sklearn.model_selection import train_test_split
x = scaledx
y = dt['AQI']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
li = LinearRegression()
li.fit(x_train,y_train)
y_pred = li.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
r2_kn = r2_score(y_test,knn.predict(x_test))
r2_d = r2_score(y_test,dt1.predict(x_test))
r2_rf = r2_score(y_test, reg.predict(x_test))
r2_lr = r2_score(y_test, li.predict(x_test))
print("R² Score - Kneighbour:", r2_kn)
print("R² Score - DecisionTree:", r2_d)
print("R² Score - Random Forest:", r2_rf)
print("R² Score - Linear Regression:", r2_lr)


models = ['KNeighborsRegressor','DecisionTree','Random Forest', 'Linear Regression']
r2_scores = [r2_kn,r2_d,r2_rf, r2_lr]
plt.figure(figsize=(8, 5))
plt.bar(models, r2_scores, color=['forestgreen', 'royalblue','red','orange'])
plt.title('Model Comparison - R² Score')
plt.ylabel('R² Score')
plt.ylim(0, 1)  # Optional: set y-axis limit between 0 and 1
for i, v in enumerate(r2_scores):
    plt.text(i, v + 0.01, f"{v:.2f}", ha='center', fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
